Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
data = a.T
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')

X = data[0]
H, L, C = data[1], data[2], data[3]

plt.plot(X, H, label='Hare', c = 'b', linewidth = 2)
plt.plot(X, L, label='Lynx', c='r', linewidth = 2)
plt.plot(X, C, label='Carrots', c = 'orange', linewidth = 2)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Population')
plt.xticks(np.arange(min(X), max(X)+1, 2.0))
plt.legend()
plt.show()
```

Find which years any of the populations is above 50000. Show time periods where populations are above 50000
in the plot of populations.

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
data = a.T
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')

X = data[0]
H, L, C = data[1], data[2], data[3]

plt.plot(X, H, label='Hare', c = 'b', linewidth = 2)
plt.plot(X, L, label='Lynx', c='r', linewidth = 2)
plt.plot(X, C, label='Carrots', c = 'orange', linewidth = 2)
plt.plot(X, np.ones(len(X)) * 50000, label = '50000', c = 'g', linestyle='--')
plt.xlabel('Year', fontsize=16)
plt.ylabel('Population')
plt.xticks(np.arange(min(X), max(X)+1, 2.0))

Hmask = [H > 50000]
Lmask = [L > 50000]
Cmask = [C > 50000]
plt.scatter(X[Hmask], H[Hmask], color = 'g')
plt.scatter(X[Lmask], L[Lmask], color = 'g')
plt.scatter(X[Cmask], C[Cmask], color = 'g')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
data = a.T
fig = plt.figure()
ax = plt.axes()
fig.patch.set_facecolor('xkcd:white')

X = data[0]
H, L, C = data[1], data[2], data[3]

# 
total = H + L
proportional_H = H / total * 100
proportional_L = L / total * 100
ax.bar(X, proportional_H, 0.5, label='Hare')
ax.bar(X, proportional_L, 0.5, bottom=proportional_H, label='Lynxes')

plt.xlabel('Year', fontsize=16)
plt.ylabel('Population')
plt.xticks(np.arange(min(X), max(X)+1, 2.0))

plt.legend()
plt.show()
```

Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
import numpy as np
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 3, sharey=True)

axs[0].barh(X, H, label='Hare')
axs[0].vlines(np.mean(H), 1899, 1921, colors = 'r')
axs[0].vlines(np.std(H), 1899, 1921, colors = 'm')
axs[0].legend()

axs[1].barh(X, L, label='Lynx')
axs[1].vlines(np.mean(L), 1899, 1921, colors = 'r')
axs[1].vlines(np.std(L), 1899, 1921, colors = 'm')
axs[1].legend()

axs[2].barh(X, C, label='Carrot')
axs[2].vlines(np.mean(C), 1899, 1921, colors = 'r')
axs[2].vlines(np.std(C), 1899, 1921, colors = 'm')
axs[2].legend()

axs[0].grid()
axs[1].grid()
axs[2].grid()
plt.show()

print("Mean of Hares:   ", round(np.mean(H), 2), " Std of Hares:   ", round(np.std(H), 2))
print("Mean of Lynxes:  ", round(np.mean(L), 2), " Std of Lynxes:  ", round(np.std(L), 2))
print("Mean of Carrots: ", round(np.mean(C), 2), "  Std of Carrots: ", round(np.std(C), 2))
```

Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

```python
more = np.where(H - L > 0, 'H', 'L')
count_H = len(np.argwhere(more == 'H'))
count_L = len(np.argwhere(more == 'L'))
print(more)

plt.pie([count_H, count_L], labels = (['Hare', 'Lynx']), autopct='%1.2f%%')
plt.show()
```
